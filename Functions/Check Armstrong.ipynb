Write a Program to determine if the given number is Armstrong number or not. Print true if number is armstrong, otherwise print false.

An Armstrong number is a number (with digits n) such that the sum of its digits raised to nth power is equal to the number itself.


For example,
371, as 3^3 + 7^3 + 1^3 = 371
1634, as 1^4 + 6^4 + 3^4 + 4^4 = 1634

```python
## Read input as specified in the question.
## Print output as specified in the question.
def check_armstrong(num):
    count = 0
    temp = num
    copy = num
    digit = 0
    flag = False
    sum = 0
    while (num>0):
        num = num//10
        count = count+1

    while(temp!=0):
        digit = temp%10
        sum = sum + digit ** count
        temp = temp//10
    if sum == copy:
        flag = True
    else:
        flag = False
    return flag    
    

n = int(input())
ans = check_armstrong(n)
if ans:
    print("true")
else:
    print("false")
    

#Alternative approach:
# def count_num(num):
#     count = 0
#     while (num>0):
#         num = num//10
#         count = count+1
#     return count    

# def check_armstrong(num):
#     a = count_num(num)
#     temp = num
#     digit = 0
#     sum = 0
#     flag = False
#     while(num!=0):
#         digit = num%10
#         sum = sum + digit ** a
#         num = num//10
#     if  sum == temp:
#         flag = True
#     else:
#         flag = False
#     return flag

# n = int(input())
# ans = check_armstrong(n)
# if ans:
#     print("true")
# else:
#     print("false")

    
    
    
```
