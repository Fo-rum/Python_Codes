Print the following pattern for the given number of rows.
Note: N is always odd.




```python
## Read input as specified in the question.
## Print output as specified in the question.
n = int(input())
n1 = (n+1)/2
n2 = n1-1

#printing first half of the pattern
i = 1
while i<=n1:
    spaces = 1
    while spaces <= n1-i:
        print(' ', end='')
        spaces = spaces + 1
    stars = 1
    while stars <= 2*i-1:
        print('*', end='')
        stars = stars + 1
    print()
    i = i+1
#printing second half of the pattern  
k = n2
while k>0:
    spaces = 1
    while spaces <= n2-k+1:
        print(' ', end='')
        spaces = spaces + 1
    stars = 1
    while stars <= 2*k-1:
        print('*', end='')
        stars = stars + 1
    print()
    k = k - 1

    
```
